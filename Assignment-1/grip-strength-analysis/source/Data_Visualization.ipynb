```python
# Create a CSV file for the raw data
import pandas as pd

# Data for the participants
data = {
    'Height': [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    'Weight': [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    'Age': [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    'Grip strength': [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    'Frailty': ['N', 'N', 'N', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'N']
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('raw_data.csv', index=False)
```

```python
# Data Cleaning Script
import pandas as pd

# Load the raw data
df = pd.read_csv('raw_data.csv')

# Check for missing values and duplicates
print(df.isnull().sum())
print(df.duplicated().sum())

# Remove duplicates if any
df = df.drop_duplicates()

# Save the cleaned data
df.to_csv('cleaned_data.csv', index=False)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
data = pd.read_csv("cleaned_data.csv")

# Correlation analysis
# Convert the 'Frailty' column to numeric (0 for 'N', 1 for 'Y')
data['Frailty_numeric'] = data['Frailty'].map({'N': 0, 'Y': 1})
correlation = data['Grip strength'].corr(data['Frailty_numeric'])
print(f"Correlation between Grip Strength and Frailty: {correlation:.2f}")

# Set the visualisation style
sns.set(style="whitegrid")

# 1. Histogram of Grip Strength
plt.figure(figsize=(10, 6))
sns.histplot(data['Grip strength'], bins=10, color='blue', edgecolor='black')
plt.title("Histogram of Grip Strength")
plt.xlabel("Grip Strength (kg)")
plt.ylabel("Frequency")
plt.show()

# 2. Box Plot of Grip Strength by Frailty
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty', y='Grip strength', data=data, palette='pastel')
plt.title("Box Plot of Grip Strength by Frailty")
plt.xlabel("Frailty")
plt.ylabel("Grip Strength (kg)")
plt.show()
```
